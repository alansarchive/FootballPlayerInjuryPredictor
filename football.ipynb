# Data Preprocessing

```python
data['Position'] = data['Position'].astype('category')
```

```python
data.describe().T
```

```python
injury_data = data.groupby("Injury_Next_Season").mean(numeric_only=True).round(2)
injury_data
```

```python
import seaborn as sns
sns.pairplot(
    data[["Sleep_Hours_Per_Night", "Nutrition_Quality_Score", "Stress_Level_Score", "Injury_Next_Season", "Knee_Strength_Score"]],
    hue = "Injury_Next_Season",
    height = 4,
    palette = "Set1"
)
```

```python
data = pd.get_dummies(data)
```

# Logistic Regression Model Build

```python
data.columns
```

```python
y = data["Injury_Next_Season"]
```

```python
x = data.drop("Injury_Next_Season", axis = 1)
```

```python
from sklearn.model_selection import train_test_split

```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y ,
    train_size = 0.8, 
    random_state = 1
)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(penalty=None, solver = 'newton-cg', max_iter = 150)
```

```python
lr.fit(x_train, y_train)
```

### Model accuracy score

```python
round(lr.score(x_train, y_train), 3)
```

```python
round(lr.score(x_test, y_test), 3)
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

pred = lr.predict(x_test)
ypred = list(map(round, pred))
cm = confusion_matrix(y_test, ypred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
